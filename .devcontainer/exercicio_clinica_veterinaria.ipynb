```python
dados_mongo = list(colecao.find({
    'cidade': {'$in': ['Assis', 'Ourinhos', 'Pompeia', 'Marília']}
}))

display(dados_mongo[:3])
```

```python
df = pd.DataFrame(dados_mongo)
df['_id'] = df['_id'].astype(str)
df.head(3)
```

```python
print(df.isnull().sum())
```

```python
faturamento_total = df['valor'].sum()
print(f"Faturamento Total: R$ {faturamento_total:.2f}")
```

```python
faturamento_por_cidade = df.groupby('cidade')['valor'].sum().sort_values(ascending=False)
faturamento_por_cidade
```

```python
faturamento_por_servico = df.groupby('servico')['valor'].sum().sort_values(ascending=False)
faturamento_por_servico
```

```python
ticket_medio_geral = df['valor'].mean()
print(f"Ticket Médio Geral: R$ {ticket_medio_geral:.2f}")
```

```python
ticket_medio_por_cidade = df.groupby('cidade')['valor'].mean().round(2)
ticket_medio_por_cidade
```

```python
servico_mais_lucrativo = df.groupby('servico')['valor'].sum().idxmax()
print(f"Serviço mais lucrativo: {servico_mais_lucrativo}")
```

```python
cidade_maior_faturamento = df.groupby('cidade')['valor'].sum().idxmax()
print(f"Cidade com maior faturamento: {cidade_maior_faturamento}")
```

```python
top_3 = df.nlargest(3, 'valor')[['tutor', 'animal', 'servico', 'valor', 'cidade']]
top_3
```

```python
top_3_lista = df.nlargest(3, 'valor')[['tutor', 'animal', 'servico', 'valor', 'cidade']].to_dict(orient='records')

relatorio = {
    'data_geracao': datetime.now(),
    'faturamento_total': float(faturamento_total),
    'faturamento_por_cidade': faturamento_por_cidade.to_dict(),
    'faturamento_por_servico': faturamento_por_servico.to_dict(),
    'ticket_medio_geral': round(float(ticket_medio_geral), 2),
    'ticket_medio_por_cidade': ticket_medio_por_cidade.to_dict(),
    'servico_mais_lucrativo': servico_mais_lucrativo,
    'cidade_maior_faturamento': cidade_maior_faturamento,
    'top_3_atendimentos': top_3_lista
}

display(relatorio)
```
